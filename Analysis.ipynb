```python
# Assuming 'data' is your DataFrame and it has columns 'category' and 'value'
import matplotlib.pyplot as plt
movies_df = data[(data['Category'] == 'TV Show') ]
grouped_df = movies_df.groupby(['Country', 'Year']).size().reset_index(name='count')

grouped_df = grouped_df.sort_values('count', ascending=False).head(10)
```

```python

sns.barplot(x='Year', y='count', hue='Country', data=grouped_df)
plt.title('Max Number Of TV Show released by the country')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Country', loc='upper left')
plt.show()
```
